```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
from PIL import Image
```

```python
def create_black_hole(radius, size, strength):
    x = np.linspace(-size, size, 2 * size)
    y = np.linspace(-size, size, 2 * size)
    X, Y = np.meshgrid(x, y)
    
    R = np.sqrt(X**2 + Y**2)
    
    mask = R < radius
    
    image = np.zeros((2 * size, 2 * size))
    image[mask] =strength
    
    blurred_image = gaussian_filter(image, sigma = radius / 2)
    
    return blurred_image
```

```python
radius = 500
size = 512
strength = 255
```

```python
black_hole_image = create_black_hole(radius, size, strength)

black_hole_image = black_hole_image.astype(np.uint8)
```

```python
plt.imshow(black_hole_image, cmap = 'gray')
plt.axis('off')
plt.show()
```

```python
black_hole_pil_image = Image.fromarray(black_hole_image)
black_hole_pil_image.save('black_hole.png')
```
